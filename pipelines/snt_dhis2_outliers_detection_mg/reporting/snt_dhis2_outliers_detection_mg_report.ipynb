# MG Outliers Detection Report

Quick reporting placeholder for the dedicated MG pipeline.

```r
# Setup
ROOT_PATH <- "~/workspace"
CODE_PATH <- file.path(ROOT_PATH, "code")
DATA_PATH <- file.path(ROOT_PATH, "data", "dhis2", "outliers_detection")
CONFIG_PATH <- file.path(ROOT_PATH, "configuration")

source(file.path(CODE_PATH, "snt_utils.r"))
install_and_load(c("jsonlite", "arrow", "glue", "reticulate"))

# Align logging init with other production notebooks
Sys.setenv(RETICULATE_PYTHON = "/opt/conda/bin/python")
openhexa <- reticulate::import("openhexa.sdk")

config_json <- fromJSON(file.path(CONFIG_PATH, "SNT_config.json"))
COUNTRY_CODE <- config_json$SNT_CONFIG$COUNTRY_CODE

# Quick checks
partial_fp <- file.path(DATA_PATH, paste0(COUNTRY_CODE, "_outlier_magic_glasses_partial.parquet"))
complete_fp <- file.path(DATA_PATH, paste0(COUNTRY_CODE, "_outlier_magic_glasses_complete.parquet"))

if (file.exists(partial_fp)) {
  partial_tbl <- arrow::read_parquet(partial_fp)
  log_msg(glue("MG partial outliers: {nrow(partial_tbl)} rows"))
} else {
  log_msg("MG partial outlier file not found.")
}

if (file.exists(complete_fp)) {
  complete_tbl <- arrow::read_parquet(complete_fp)
  log_msg(glue("MG complete outliers: {nrow(complete_tbl)} rows"))
} else {
  log_msg("MG complete outlier file not found.")
}
```
