# DHIS2 formatting — rapport

Complétude HF et par district, cohérence des indicateurs (agrégations **ADM2×mois**), cartes et nuages population. Paramètres Papermill : fenêtre d’années, fenêtre mois pour les graphiques de complétude, plafond d’indicateurs pour les pivots, sous-échantillonnage des nuages, simplification des géométries.

**Dépendances workspace :** `code/snt_utils.r`, `pipelines/snt_dhis2_formatting/utils/snt_dhis2_formatting_report.r`, `pipelines/snt_dhis2_formatting/utils/snt_dhis2_formatting.r`. **Ordre d’exécution :** chemins → `snt_utils` → paquets + `source(..., snt_dhis2_formatting_report.r)` → OpenHEXA → configuration → données.

**Exécution :** Jupyter / Papermill.

## 1. Complétude (résumé HF / district) — fenêtre récente + indicateurs plafonnés

```r
data <- routine_data %>% dplyr::mutate(DATE = lubridate::ymd(paste0(PERIOD, "01")))
indicator_vars <- setdiff(names(data), c(
  "PERIOD", "YEAR", "MONTH", "OU_ID", "OU_NAME", "ADM1_NAME", "ADM1_ID", "ADM2_NAME", "ADM2_ID", "DATE"
))
if (length(indicator_vars) == 0L) {
  message("Aucune colonne indicateur après setdiff — section 1.1 omise.")
} else {
long_data <- data %>%
  tidyr::pivot_longer(dplyr::all_of(indicator_vars), names_to = "INDICATOR", values_to = "VALUE") %>%
  dplyr::rename(OU = OU_ID)
long_data <- report_limit_long_indicators(long_data, "INDICATOR", REPORT_MAX_INDICATORS)
reporting_summary <- reporting_summary_facility_chunked(long_data)
report_release_objects("long_data", "data", "indicator_vars", full_gc = TRUE)
plot_data <- reporting_summary %>%
  tidyr::pivot_longer(dplyr::starts_with("pct_"), names_to = "Status", values_to = "Percentage") %>%
  dplyr::mutate(Status = dplyr::recode(Status, pct_missing = "Valeur manquante",
    pct_zero = "Valeur nulle rapportée", pct_positive = "Valeur positive rapportée")) %>%
  tidyr::complete(INDICATOR, DATE, Status, fill = list(Percentage = 0))
plot_data <- report_filter_recent_months(plot_data, "DATE", REPORT_PLOT_MONTHS)
run_plot_guarded("1.1 complétude HF", function() {
  if (skip_heavy_plot_input(plot_data, "1.1 plot_data")) return(invisible(NULL))
  ncol_use <- 4L
  n_ind <- dplyr::n_distinct(plot_data$INDICATOR)
  n_row <- as.integer(ceiling(n_ind / ncol_use))
  repr_h <- max(12, min(36, 6 + 1.35 * n_row))
  options(repr.plot.width = 20, repr.plot.height = repr_h, repr.plot.res = 120)
  p <- ggplot2::ggplot(plot_data, ggplot2::aes(x = DATE, y = Percentage, fill = Status)) +
    ggplot2::geom_col(position = "stack") +
    ggplot2::facet_wrap(~ INDICATOR, scales = "free_y", ncol = 4) +
    ggplot2::scale_y_continuous() +
    ggplot2::scale_fill_manual(values = c(
      "Valeur manquante" = "tomato", "Valeur nulle rapportée" = "skyblue",
      "Valeur positive rapportée" = "green"
    )) +
    ggplot2::labs(
      title = "Taux de rapportage par indicateur (niveau formation sanitaire)",
      subtitle = "Proportion des valeurs rapportées par mois et par indicateur",
      x = "Mois", y = "% des formations sanitaires", fill = "Statut du rapportage"
    ) +
    ggplot2::theme_minimal(base_size = 16) +
    ggplot2::theme(
      plot.title = ggplot2::element_text(face = "bold", size = 20),
      strip.text = ggplot2::element_text(size = 16),
      axis.title = ggplot2::element_text(size = 16),
      axis.text = ggplot2::element_text(size = 16)
    )
  print(p)
})
report_release_objects("plot_data", "reporting_summary", full_gc = TRUE)
}
```

```r
data <- routine_data %>% dplyr::mutate(Date = lubridate::ymd(paste0(PERIOD, "01")))
indicator_cols <- setdiff(names(data), c(
  "PERIOD", "YEAR", "MONTH", "OU_ID", "OU_NAME", "ADM1_NAME", "ADM1_ID", "ADM2_NAME", "ADM2_ID", "Date"
))
if (length(indicator_cols) == 0L) {
  message("Aucune colonne indicateur — section 1.2 omise.")
} else {
if (length(indicator_cols) > REPORT_MAX_INDICATORS) indicator_cols <- indicator_cols[seq_len(REPORT_MAX_INDICATORS)]
data_long <- data %>%
  dplyr::select(ADM2_ID, OU_ID, Date, dplyr::all_of(indicator_cols)) %>%
  tidyr::pivot_longer(dplyr::all_of(indicator_cols), names_to = "Indicator", values_to = "value") %>%
  dplyr::mutate(value = as.numeric(value))
reporting_summary <- reporting_summary_adm2_chunked(data_long)
plot_data <- reporting_summary %>%
  tidyr::pivot_longer(dplyr::starts_with("pct_"), names_to = "Status", values_to = "Percentage") %>%
  dplyr::mutate(Status = dplyr::recode(Status, pct_missing = "Valeur manquante",
    pct_zero = "Valeur nulle rapportée", pct_positive = "Valeur positive rapportée")) %>%
  tidyr::complete(Indicator, Date, Status, fill = list(Percentage = 0))
plot_data <- report_filter_recent_months(plot_data, "Date", REPORT_PLOT_MONTHS)
run_plot_guarded("1.2 complétude district", function() {
  if (skip_heavy_plot_input(plot_data, "1.2 plot_data")) return(invisible(NULL))
  n_ind <- dplyr::n_distinct(plot_data$Indicator)
  ncol_guess <- max(1L, as.integer(ceiling(sqrt(n_ind))))
  n_row <- as.integer(ceiling(n_ind / ncol_guess))
  repr_h <- max(12, min(36, 6 + 1.28 * n_row))
  options(repr.plot.width = 20, repr.plot.height = repr_h, repr.plot.res = 120)
  p <- ggplot2::ggplot(plot_data, ggplot2::aes(x = Date, y = Percentage, fill = Status)) +
    ggplot2::geom_col(position = "stack") +
    ggplot2::facet_wrap(~ Indicator, scales = "free_y") +
    ggplot2::scale_y_continuous(limits = c(0, 100)) +
    ggplot2::scale_fill_manual(values = c(
      "Valeur manquante" = "tomato", "Valeur nulle rapportée" = "skyblue",
      "Valeur positive rapportée" = "green"
    )) +
    ggplot2::labs(
      title = "Taux de rapportage par indicateur (niveau district)",
      subtitle = "Proportion des districts (ADM2_ID) rapportant chaque mois",
      x = "Mois", y = "% des districts", fill = "Statut du rapportage"
    ) +
    ggplot2::theme_minimal(base_size = 14) +
    ggplot2::theme(
      plot.title = ggplot2::element_text(face = "bold", size = 18),
      strip.text = ggplot2::element_text(size = 14),
      axis.title = ggplot2::element_text(size = 14),
      axis.text = ggplot2::element_text(size = 12)
    )
  print(p)
})
report_release_objects("data_long", "data", "plot_data", "reporting_summary", full_gc = TRUE)
}
```

## 2. Cohérence des indicateurs

Agrégation **ADM2 × mois** sur la fenêtre d’années chargée. **Courbe nationale :** construction des mois à partir de PERIOD (année + mois → date du 1er du mois) ; les **derniers mois dont les totaux nationaux sont entièrement à 0** (extrait souvent incomplet en fin de série, `sum(..., na.rm=TRUE)` quand toutes les valeurs OU sont NA) sont **retirés** pour éviter une chute artificielle des courbes.

```r
routine_hd_month <- routine_adm2_month_from_wide(routine_data)
routine_hd_month <- report_downsample_rows(routine_hd_month, 12000L)
run_plot_guarded("2.2 nuages ADM2-mois", function() {
  if (nrow(routine_hd_month) == 0L) return(invisible(NULL))
  if (!all(c("SUSP", "TEST", "CONF") %in% names(routine_hd_month))) {
    message("Colonnes SUSP/TEST/CONF manquantes après agrégation — nuages omis.")
    return(invisible(NULL))
  }
  if (skip_heavy_plot_input(routine_hd_month, "2.2 routine_hd_month")) return(invisible(NULL))
  options(repr.plot.width = 18, repr.plot.height = 8, repr.plot.res = 120)
  p1 <- ggplot2::ggplot(routine_hd_month, ggplot2::aes(x = SUSP, y = TEST)) + ggplot2::geom_point(alpha = 0.5, color = "blue") +
    ggplot2::geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") +
    ggplot2::labs(title = "Suspectés vs Testés", x = "Cas suspectés", y = "Cas testés") + ggplot2::theme_minimal(base_size = 16)
  p2 <- ggplot2::ggplot(routine_hd_month, ggplot2::aes(x = TEST, y = CONF)) + ggplot2::geom_point(alpha = 0.5, color = "darkgreen") +
    ggplot2::geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") +
    ggplot2::labs(title = "Testés vs Confirmés", x = "Cas testés", y = "Cas confirmés") + ggplot2::theme_minimal(base_size = 16)
  if (all(c("CONF", "MALTREAT") %in% names(routine_hd_month))) {
    p3 <- ggplot2::ggplot(routine_hd_month, ggplot2::aes(x = CONF, y = MALTREAT)) + ggplot2::geom_point(alpha = 0.5, color = "purple") +
      ggplot2::geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") +
      ggplot2::labs(title = "Confirmés vs Traités", x = "Cas confirmés", y = "Cas traités") + ggplot2::theme_minimal(base_size = 16)
  } else {
    p3 <- ggplot2::ggplot() + ggplot2::theme_void() + ggplot2::labs(title = "MALTREAT absent")
  }
  print((p1 | p2 | p3) + patchwork::plot_layout(guides = "collect"))
})
rds_clean_month <- routine_national_month_from_wide(routine_data)
run_plot_guarded("2.2 tendances nationales", function() {
  if (nrow(rds_clean_month) == 0L) return(invisible(NULL))
  if (skip_heavy_plot_input(rds_clean_month, "2.2 rds_clean_month")) return(invisible(NULL))
  vars_nat <- intersect(c("SUSP", "TEST", "CONF", "PRES"), names(rds_clean_month))
  if (length(vars_nat) == 0L) return(invisible(NULL))
  options(repr.plot.width = 16, repr.plot.height = 8, repr.plot.res = 120)
  p <- rds_clean_month %>%
    tidyr::pivot_longer(dplyr::all_of(vars_nat), names_to = "Indicator") %>%
    ggplot2::ggplot(ggplot2::aes(x = DATE, y = value, color = Indicator)) +
    ggplot2::geom_line(linewidth = 1.2) +
    ggplot2::labs(
      title = "Tendances mensuelles nationales des indicateurs composites",
      x = "Mois", y = "Nombre de cas", color = "Indicateur"
    ) +
    ggplot2::theme_minimal(base_size = 16) +
    ggplot2::theme(
      plot.title = ggplot2::element_text(face = "bold", size = 20),
      axis.title = ggplot2::element_text(size = 16),
      axis.text = ggplot2::element_text(size = 16),
      legend.title = ggplot2::element_text(size = 16),
      legend.text = ggplot2::element_text(size = 16)
    )
  print(p)
})
report_release_objects("routine_hd_month", "rds_clean_month", "routine_data", full_gc = TRUE)
```

## 3. Cartes population ADM2 et complétude population

```r
if (!is.null(population_data) && !is.null(shapes_data)) {
  map_data <- dplyr::left_join(shapes_data, population_data, by = "ADM2_ID")
  run_plot_guarded("3.1 carte population", function() {
    if (skip_heavy_plot_input(map_data, "3.1 map_data")) return(invisible(NULL))
    options(repr.plot.width = 20, repr.plot.height = 12, repr.plot.res = 120)
    p <- ggplot2::ggplot(map_data) +
      ggplot2::geom_sf(ggplot2::aes(fill = POPULATION), color = "white", linewidth = 0.2) +
      ggplot2::scale_fill_viridis_c(option = "C", name = "Population") +
      ggplot2::labs(
        title = "Population totale par district sanitaire (DS)",
        subtitle = "Données DHIS2",
        caption = "Source: NMDR / DHIS2"
      ) +
      ggplot2::theme_minimal(base_size = 14)
    print(p)
  })
  rm(map_data)
} else print("Population ou shapes indisponibles.")
```

```r
population_data_filtered <- population_data
if (identical(COUNTRY_CODE_CHR, "NER") && !is.null(population_data)) {
  years_to_keep <- 2022:2024
  population_data_filtered <- dplyr::filter(population_data, YEAR %in% years_to_keep)
  metadata_json <- jsonlite::fromJSON(file.path(CONFIG_PATH, "SNT_metadata.json"))
  value_breaks_tot <- jsonlite::fromJSON(metadata_json$POPULATION_TOTAL$SCALE)
  value_breaks_u5 <- jsonlite::fromJSON(metadata_json$POPULATION_U5$SCALE)
  value_breaks_fe <- jsonlite::fromJSON(metadata_json$POPULATION_PREGNANT$SCALE)
  labels_tot <- create_dynamic_labels(value_breaks_tot)
  labels_u5 <- create_dynamic_labels(value_breaks_u5)
  labels_fe <- create_dynamic_labels(value_breaks_fe)
  NER_palette_population <- c("1" = "#fae6db", "2" = "#f1b195", "3" = "#ea7354", "4" = "#cc3f32", "5" = "#972620")
} else {
  labels_tot <- labels_u5 <- labels_fe <- NULL
  NER_palette_population <- NULL
}
```

```r
if (identical(COUNTRY_CODE_CHR, "NER") && !is.null(population_data_filtered) && !is.null(shapes_data)) {
  if (!exists("NER_palette_population") || is.null(NER_palette_population) || is.null(labels_tot)) {
    message("NER: palette ou labels absents — exécuter la cellule metadata NER juste au-dessus.")
  } else if (length(NER_palette_population) != length(labels_tot)) {
    warning("NER: longueur palette (", length(NER_palette_population), ") != labels_tot (", length(labels_tot), ") — choropleth omis.")
  } else {
    names(NER_palette_population) <- labels_tot
    run_plot_guarded("NER choropleth population totale", function() {
    options(repr.plot.width = 20, repr.plot.height = 12, repr.plot.res = 120)
    plot <- population_data_filtered %>% dplyr::mutate(CATEGORY_POPULATION = cut(POPULATION,
      breaks = c(0, value_breaks_tot, Inf), labels = labels_tot, right = TRUE, include.lowest = TRUE)) %>%
      dplyr::left_join(shapes_data, by = dplyr::join_by(ADM1_NAME, ADM1_ID, ADM2_NAME, ADM2_ID)) %>%
      ggplot2::ggplot() + ggplot2::geom_sf(ggplot2::aes(geometry = geometry, fill = CATEGORY_POPULATION), linewidth = 0.15) +
      ggplot2::facet_wrap(~YEAR, ncol = 3) + ggplot2::theme_void() + ggplot2::scale_fill_manual(values = NER_palette_population, limits = labels_tot, drop = FALSE) +
      ggplot2::labs(title = "Population totale (NER)", subtitle = "Source: NMDR / DHIS2")
    print(plot)
      ggplot2::ggsave(file.path(figures_dir, paste0(COUNTRY_CODE, "_choropleth_population_totals.png")), plot, width = 12, height = 7, dpi = REPORT_FIG_DPI)
    })
  }
}
```

```r
if (!is.null(population_data_filtered)) {
  pop_scatter <- report_downsample_rows(population_data_filtered, REPORT_POP_SCATTER_MAX_ROWS)
  run_plot_guarded("Population totale vs district", function() {
    if (skip_heavy_plot_input(pop_scatter, "3.2 population scatter")) return(invisible(NULL))
    n_adm1 <- dplyr::n_distinct(pop_scatter$ADM1_NAME)
    repr_h <- max(26, min(42, 12 + 2.8 * n_adm1))
    options(repr.plot.width = 20, repr.plot.height = repr_h, repr.plot.res = 120)
    p <- ggplot2::ggplot(pop_scatter) +
      ggplot2::geom_point(ggplot2::aes(x = POPULATION, y = forcats::fct_reorder(ADM2_NAME, POPULATION), color = factor(YEAR)), size = 0.5, alpha = 0.65) +
      ggplot2::facet_grid(rows = ggplot2::vars(ADM1_NAME), scales = "free_y", space = "free_y", switch = "y") +
      ggplot2::scale_x_continuous(labels = scales::comma) +
      ggplot2::labs(color = "Année") +
      ggplot2::theme_minimal() +
      ggplot2::theme(
        axis.text = ggplot2::element_text(size = 7),
        axis.title.x = ggplot2::element_text(size = 7),
        axis.title.y = ggplot2::element_blank(),
        strip.placement = "outside",
        panel.grid.minor.x = ggplot2::element_blank(),
        legend.position = "bottom"
      )
    print(p)
    ggplot2::ggsave(file.path(figures_dir, "hist_population_totale.png"), p, width = 15, height = 23, units = "cm", dpi = REPORT_FIG_DPI, bg = "white")
  })
}
```
